```python
import random

# Define questions and answers
questions = {
    "Height_of_Kome?": ["A. 170 cm", "B. 160 cm", "C. 175 cm", "D. 177 cm"],
    "Height_of_Victoria?": ["A. 164 cm", "B. 168 cm", "C. 175 cm", "D. 179 cm"],
    "What is the chemical symbol for water?": ["A. H2O", "B. CO2", "C. O2", "D. NaCl"],
    "Club of Mbappe?": ["A. PSG", "B. Chelsea", "C. Liverpool", "D. Julius Berger"],
    "Club of Victor Osimhen?": ["A. Napoli", "B. Enyimba", "C. PSG", "D. Inter Miami"]
}

# Define answers
answers = {
    "Height_of_Kome?": "A",
    "Height_of_Victoria?": "A",
    "What is the chemical symbol for water?": "A",
    "Club of Mbappe?": "A",
    "Club of Victor Osimhen?": "A"
}

# Function to display and shuffle questions
def display_questions():
    shuffled_questions = list(questions.keys())
    random.shuffle(shuffled_questions)
    score = 0
    
    for question in shuffled_questions:
        print(question)
        for option in questions[question]:
            print(option)
        user_answer = input("Enter your answer (A/B/C/D): ").upper()
        if user_answer == answers[question]:
            print("Correct!")
            score += 1
        else:
            print("Incorrect!")
    
    print("Quiz completed! You scored {}/{}.".format(score, len(questions)))

# Main function
def main():
    print("Welcome to the Quiz Game!")
    print("You will be asked unrelated random questions from various topics. Please select the correct option (A, B, C, or D).")
    display_questions()

if __name__ == "__main__":
    main()
   
```

```python
import random

# Define questions for each category
categories = {
    "Geography": [
        {
            "question": "What is the capital of Morocco?",
            "options": ["Rabat", "Casablanca", "Mecca", "Maghreb"],
            "answer": "Rabat"
        },
        {
            "question": "What is the capital of Australia?",
            "options": ["Melbourne", "Sydney", "Canberra", "Adelaide"],
            "answer": "Canberra"
        },
        {
            "question": "What is the executive capital of South Africa?",
            "options": ["Johannesburg", "Pretoria", "Bloemfontein", "Cape Town"],
            "answer": "Pretoria"
        },
        {
            "question": "What is the capital of Canada?",
            "options": ["Toronto", "Ontario", "Ottawa", "Quebec"],
            "answer": "Ottawa"
        }
    ],
    
    "Sports": [
        {
            "question": "What is the current soccer club of Mbappe?",
            "options": ["Chelsea", "PSG", "Liverpool", "Julius Berger"],
            "answer": "PSG"
        },
        {
            "question": "What is the current soccer club of Messi?",
            "options": ["Inter Miami CF", "PSG", "Kano Pillars", "Julius Berger"],
            "answer": "Inter Miami CF"
        },
        {
            "question": "What is the current soccer club of Victor Osimhen?",
            "options": ["Wikki Tourists F.C", "PSG", "Napoli", "Julius Berger"],
            "answer": "Napoli"
        },
        {
            "question": "What is the current soccer club of Haaland?",
            "options": ["Villarreal CF", "PSG", "Manchester City", "Julius Berger"],
            "answer": "Manchester City"
        }
    ]
}

def quiz():
    print("Welcome to the Quiz Game!")
    while True:
        print("\nChoose a category:")
        for i, category in enumerate(categories.keys()):
            print(f"{i+1}. {category}")
        print("0. Quit")
        
        choice = input("Enter your choice: ")
        
        if choice == '0':
            print("Thank you for playing!")
            break
        
        try:
            choice = int(choice)
            if choice not in range(1, len(categories) + 1):
                print("Invalid choice! Please choose again.")
                continue
            
            category = list(categories.keys())[choice - 1]
            questions = categories[category]
            random.shuffle(questions)
            
            score = 0
            for question in questions:
                print("\n" + question["question"])
                for i, option in enumerate(question["options"]):
                    print(f"{i+1}. {option}")
                answer = input("Enter your answer (1-4): ")
                if answer == str(question["options"].index(question["answer"]) + 1):
                    print("Correct!")
                    score += 1
                else:
                    print(f"Wrong! The correct answer is: {question['answer']}")
            
            print(f"\nYour score for {category} category: {score}/{len(questions)}")
            
        except ValueError:
            print("Invalid choice! Please enter a number.")
            

if __name__ == "__main__":
    quiz()
```

```python
def display_menu():
    print("Todo List Manager")
    print("1. View Todo List")
    print("2. Add New Todo")
    print("3. Mark Todo as Completed")
    print("4. Remove Todo")
    print("5. Quit")

def view_todo_list(todo_list):
    print("Todo List:")
    if not todo_list:
        print("No todos.")
    else:
        for i, todo in enumerate(todo_list, start=1):
            print(f"{i}. {todo}")

def add_todo(todo_list):
    new_todo = input("Enter the new todo: ")
    todo_list.append(new_todo)
    print("Todo added successfully.")

def mark_completed(todo_list):
    view_todo_list(todo_list)
    try:
        index = int(input("Enter the index of the todo to mark as completed: ")) - 1
        if 0 <= index < len(todo_list):
            todo_list[index] += " - Completed"
            print("Todo marked as completed.")
        else:
            print("Invalid index.")
    except ValueError:
        print("Invalid input. Please enter a number.")

def remove_todo(todo_list):
    view_todo_list(todo_list)
    try:
        index = int(input("Enter the index of the todo to remove: ")) - 1
        if 0 <= index < len(todo_list):
            removed_todo = todo_list.pop(index)
            print(f"'{removed_todo}' removed successfully.")
        else:
            print("Invalid index.")
    except ValueError:
        print("Invalid input. Please enter a number.")

def todo_manager():
    todo_list = []
    while True:
        display_menu()
        choice = input("Enter your choice: ")
        if choice == "1":
            view_todo_list(todo_list)
        elif choice == "2":
            add_todo(todo_list)
        elif choice == "3":
            mark_completed(todo_list)
        elif choice == "4":
            remove_todo(todo_list)
        elif choice == "5":
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please choose a valid option.")

if __name__ == "__main__":
    todo_manager()
```
